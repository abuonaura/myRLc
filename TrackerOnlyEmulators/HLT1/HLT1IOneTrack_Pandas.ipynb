```python
def GetDataframes(dtype, polarity):
    i=0
    df_list=[]
    directory = 'HDF_Lb'+dtype+'_'+polarity+'/'
    if os.path.exists(filedir+directory):
        if os.path.isfile(filedir+directory+'df0.h5'):
            listfiles = os.listdir(filedir+directory) # dir is your directory path
            for i in range(len(listfiles)):
                df = pd.read_hdf(filedir+directory+'df'+str(i)+'.h5', 'df')
                df_list.append(df)
        else:
            i=0
            for df in uproot.pandas.iterate(filedir+'Lb_'+dtype+'_'+polarity+'.root','tupleout/DecayTree', entrysteps=500000):
                df_list.append(df)
                df.to_hdf(filedir+directory+'df'+str(i)+'.h5', key='df', mode='w')
                i+=1
    else:
        os.mkdir(filedir+directory)
        for df in uproot.pandas.iterate(filedir+'MC/Lb_'+dtype+'_'+polarity+'.root','tupleout/DecayTree',entrysteps=500000):
            df_list.append(df)
            df.to_hdf(filedir+directory+'df'+str(i)+'.h5', key='df', mode='w')
            i+=1
    return df_list

df_list = GetDataframes('Lcmunu','MagUp')
```

```python
#---Global event decisions
def GECDecision(df):
    df['GECDecision'] = True
    df.loc[df.nVeloClusters >=  6000,'GECDecision'] = False
    df.loc[df.nVeloClusters <=  50,'GECDecision'] = False
    df.loc[df.nOTClusters >=  15000,'GECDecision'] = False
    df.loc[df.nOTClusters <=  50,'GECDecision'] = False
    df.loc[df.nITClusters >=  3000,'GECDecision'] = False
    df.loc[df.nITClusters <=  50,'GECDecision'] = False
    return df

df1 = GECDecision(df1)
```

```python
def TrackInputDecision(df,head):
    df['TrkInDec_'+head[-1]] = True
    df.loc[df['Lb_PT_'+head] <= 500,'TrkInDec_'+head[-1]] = False
    df.loc[df['Lb_P_'+head] <= 5000,'TrkInDec_'+head[-1]] = False
    df.loc[df['Lb_TRACK_GHOSTPROB_'+head] > 999.,'TrkInDec_'+head[-1]] = False
    df.loc[df['Lb_TRACK_CHI2_'+head]/df['Lb_TRACK_NDOF_'+head] >= 4.,'TrkInDec_'+head[-1]] = False
    return df

def TrackAdditionalDecision(df,head):
    df['isAdditional_'+head[-1]] = True
    particle={'DAU_1':'K_TRACK_','DAU_2':'p_TRACK_','DAU_3':'pi_TRACK_','DAU_4':'mu_TRACK_'}
    df.loc[df[particle[head]+'nTTHits']<3,'isAdditional_'+head[-1]] = False
    return df

def TrackMVADecision(df,head):
    df['MVADecision_'+head[-1]] = True
    df.loc[(df['Lb_PT_'+head] < 500)|(df['Lb_TRACK_CHI2_'+head]>2.5),'MVADecision_'+head[-1]] = False
    df.loc[df['Lb_TRACK_GHOSTPROB_'+head] >= 0.2,'MVADecision_'+head[-1]] = False
    df.loc[(df['Lb_PT_'+head] > 25000)&(df['Lb_IPCHI2_OWNPV_'+head]>7.4),'MVADecision_'+head[-1]] = True
    df.loc[(df['Lb_PT_'+head] > 25000)&(df['Lb_IPCHI2_OWNPV_'+head]<=7.4),'MVADecision_'+head[-1]] = False
    df.loc[(df['Lb_IPCHI2_OWNPV_'+head]>0)&(df['Lb_PT_'+head]!=500)&
           (df['Lb_IPCHI2_OWNPV_'+head]<=
            1.0/pow(df['Lb_PT_'+head]/1000. - 1.0,2) + 1.1/25000.*(25000. - df['Lb_PT_'+head]) + log(7.4))
          ,'MVADecision_'+head[-1]] = False    
    return df

def Inefficiency_Decision(Correction):
        #TODO: Check how random is this generator
        rand = random.uniform(0.,1.)
        if rand > Correction: return False
        
        return True     

def TrackReconstructedDecision(df,head):
    #---Get the Track Reconstruction Efficiency correction factor
    Correction = 1./1.035 #This is a number I get from LHCb-PUB-2015-024
    isReco =  Inefficiency_Decision(Correction)
    df["isReconstructed_"+head[-1]] = True
    if isReco==0:
        df["isReconstructed_"+head[-1]] = False
    return df

def HLT1TrackMVA_EffCorr_TOS(df,head):
    df['Lc_HLT1TrackMVA_Emu_EffCorrected_TOS_'+head[-1]] = False
    df['Lc_HLT1TrackMVA_Emu_TOS_'+head[-1]] = False
    df.loc[(df['isReconstructed_'+head[-1]]) & (df['MVADecision_'+head[-1]]) & (df['TrkInDec_'+head[-1]])
               & (df['isAdditional_'+head[-1]]),'Lc_HLT1TrackMVA_Emu_EffCorrected_TOS_'+head[-1]] = True
    df.loc[(df['MVADecision_'+head[-1]]) & (df['TrkInDec_'+head[-1]])
               & (df['isAdditional_'+head[-1]]),'Lc_HLT1TrackMVA_Emu_TOS_'+head[-1]] = True
    return df
```

```python
def TwoTrackMVADecision(df,i,j):
    df['TwoTrackMVA_'+i+'_'+j] = True
    df.loc[df['Lb_VERTEX_CHI2_COMB_'+i+'_'+j] >= 10.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_VERTEX_CHI2_COMB_'+i+'_'+j] <= 0.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_ETA_COMB_'+i+'_'+j] <= 2.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_ETA_COMB_'+i+'_'+j] >= 5.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_MCORR_OWNPV_COMB_'+i+'_'+j] <= 1000.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_MCORR_OWNPV_COMB_'+i+'_'+j] >= 1000000000.0,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_MCORR_OWNPV_COMB_'+i+'_'+j] <= 0.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_DIRA_OWNPV_COMB_'+i+'_'+j] <= 0.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_HLt1TwoTrackMVAEmulations_'+i+'_'+j] <= 0.95,'TwoTrackMVA_'+i+'_'+j] = False
    particle={'1':'K','2':'p','3':'pi','4':'mu'}
    df['PX_'+i+'_'+j] = df[particle[i]+'_PX']+ df[particle[j]+'_PX']
    df['PY_'+i+'_'+j] = df[particle[i]+'_PY']+ df[particle[j]+'_PY']
    df['PZ_'+i+'_'+j] = df[particle[i]+'_PZ']+ df[particle[j]+'_PZ']
    df['PT_'+i+'_'+j] = np.sqrt(df['PX_'+i+'_'+j]*df['PX_'+i+'_'+j]+df['PY_'+i+'_'+j]*df['PY_'+i+'_'+j])
    df['P_'+i+'_'+j] = np.sqrt(df['PX_'+i+'_'+j]*df['PX_'+i+'_'+j]+df['PY_'+i+'_'+j]*df['PY_'+i+'_'+j]+
                             df['PZ_'+i+'_'+j]*df['PZ_'+i+'_'+j])
    df=df.drop(df1.loc[:, 'PX_'+i+'_'+j:'PZ_'+i+'_'+j].columns,axis=1)
    df.loc[df['PT_'+i+'_'+j] <= 2000.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_DOCA_COMB_'+i+'_'+j] <= 0.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_DOCA_COMB_'+i+'_'+j] >= 10.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_VDCHI2_OWNPV_COMB_'+i+'_'+j] <= 0.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['PT_'+i+'_'+j] <= 0.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_PT_COMB_'+i+'_'+j] <= 500.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_P_COMB_'+i+'_'+j] <= 0.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_IPCHI2_OWNPV_COMB_'+i+'_'+j] <= 4.,'TwoTrackMVA_'+i+'_'+j] = False
    df.loc[df['Lb_IPCHI2_OWNPV_COMB_'+i+'_'+j] < 0.,'TwoTrackMVA_'+i+'_'+j] = False
    return df
```

```python
def TwoTrackInputDecision(df,i):
    head = 'DAU_'+i
    df['2TrkInDec_'+i] = True
    df.loc[df['Lb_PT_'+head] <= 600,'2TrkInDec_'+i] = False
    df.loc[df['Lb_P_'+head] <= 5000,'2TrkInDec_'+i] = False
    df.loc[df['Lb_TRACK_CHI2_'+head] >= 2.5,'2TrkInDec_'+i]=False
    df.loc[df['Lb_TRACK_GHOSTPROB_'+head] >= 0.2,'2TrkInDec_'+i] = False
    df.loc[df['Lb_IPCHI2_OWNPV_'+head] <= 4.,'2TrkInDec_'+i] = False
    return df
```
