```python
import ROOT as r
```

```python
def PlotHisto(h,name):
    c = r.TCanvas(name,'',500,500)
    h.SetLineColor(r.kAzure-3)
    h.Draw('hist')
    return c

def PlotHisto2D(h,name):
    c = r.TCanvas(name,'',500,500)
    h.Draw('colz')
    return c

def PlotHistoSameCanvas(c,h,color):
    h.SetLineColor(color)
    h.Draw('hist same')
    return c

def PlotTemplates(h,name):
    h.SetLineColor(r.kAzure-3)
    c = r.TCanvas(name,'',1500,500)
    c.Divide(3,1)
    c.cd(1)
    h.ProjectionX().Draw('hist')
    c.cd(2)
    h.ProjectionY().Draw('hist')
    c.cd(3)
    h.ProjectionZ().Draw('hist')
    return c

def PlotTemplatesSameCanvas(c,h,color):
    h.SetLineColor(color)
    c.cd(1)
    h.ProjectionX().Draw('hist sames')
    c.cd(2)
    h.ProjectionY().Draw('hist sames')
    c.cd(3)
    h.ProjectionZ().Draw('hist sames')
    return c

def ScaleHisto(h,value):
    scale = value/h.Integral()
    h.Scale(scale)
    return h
```

```python
samples    = ['Lcmunu','Lctaunu','Lc2593munu','Lc2625munu','Lc2593taunu','Lc2625taunu']
polarities = ['MagUp','MagDown']
categories = ['iso','Kenr','Lcpipi']
vars2read = ['Event_FFcorr','FitVar_q2_mLc','FitVar_Mmiss2_mLc','FitVar_El_mLc','Event_PIDCalibEffWeight','w_LbCorr']
```

```python
hNoCorr = {}
hWithLbCorr = {}
hWithFF = {}
for category in categories:
    hNoCorr[category] = {}
    hWithLbCorr[category] = {}
    hWithFF[category] = {}
    for sample in samples:
        print(sample)
        hNoCorr[category][sample] = r.TH3F('hNoCorr_'+sample+'_'+category,';q^{2};E_{l};M^{2}_{miss}',4,-2,14,10,0,2600,10,-2,14)
        hWithLbCorr[category][sample] = r.TH3F('hWithLbCorr_'+sample+'_'+category,';q^{2};E_{l};M^{2}_{miss}',4,-2,14,10,0,2600,10,-2,14)
        hWithFF[category][sample] = r.TH3F('hWithFF_'+sample+'_'+category,';q^{2};E_{l};M^{2}_{miss}',4,-2,14,10,0,2600,10,-2,14)
        hNoCorr[category][sample].SetDirectory(0)
        hWithLbCorr[category][sample].SetDirectory(0)
        hWithFF[category][sample].SetDirectory(0)
        for polarity in polarities:
            print('  ',polarity)
            print(hNoCorr[category][sample])

            fname = '/disk/lhcb_data2/RLcMuonic2016/MC_full_new/Lb_'+sample+'_'+polarity+'_full_preselected_'+category+'_LbCorr.root'
            tname = 'DecayTree'
            print(hNoCorr[category][sample])
            f = r.TFile(fname, 'READ')
            print(hNoCorr[category][sample])
            t = f.Get(tname)
            
            #Switch off all branches to speed up
            #t.SetBranchStatus('*',0)
            #and turn on only those needed
            #for b in vars2read:
             #   t.SetBranchStatus(b,1)
                
            
            for i in range(t.GetEntries()):
                t.GetEntry(i)
                w = t.Event_PIDCalibEffWeight
                w_lb = t.w_LbCorr
                w_ff = t.Event_FFcorr
                
                if (t.FitVar_q2_mLc/1.E6>-2 and t.FitVar_q2_mLc/1.E6<14) and (t.FitVar_Mmiss2_mLc/1.E6>-2 and t.FitVar_Mmiss2_mLc/1.E6<14) and (t.FitVar_El_mLc>0 and t.FitVar_El_mLc<2600):
                    
                    hNoCorr[category][sample].Fill(t.FitVar_q2_mLc/1.E6,t.FitVar_El_mLc,t.FitVar_Mmiss2_mLc/1.E6,w)
                    hWithLbCorr[category][sample].Fill(t.FitVar_q2_mLc/1.E6,t.FitVar_El_mLc,t.FitVar_Mmiss2_mLc/1.E6,w*w_lb)
                    hWithFF[category][sample].Fill(t.FitVar_q2_mLc/1.E6,t.FitVar_El_mLc,t.FitVar_Mmiss2_mLc/1.E6,w*w_lb*w_ff)
```

```python
c = {}
for sample in samples:
    c[sample] = PlotTemplates(hNoCorr['iso'][sample],'c_'+sample+'_iso')
    c[sample] = PlotTemplatesSameCanvas(c[sample],hWithLbCorr['iso'][sample],r.kBlack)
    c[sample] = PlotTemplatesSameCanvas(c[sample],hWithFF['iso'][sample],r.kOrange-3)
    l = r.TLegend(0.6,0.6,0.9,0.8)
    l.AddEntry(hNoCorr['iso'][sample],'No correction','l')
    l.AddEntry(hWithLbCorr['iso'][sample],'With Lb Prod correction','l')
    l.AddEntry(hWithFF['iso'][sample],'With FF correction','l')
    l.Draw('same')
    c[sample].Draw()
```
